```python
import pandas as pd
import numpy as np

path=r"E:\Users\Boeing\Desktop\aliyun tianchi\二手车交易价格预测\analysis\\"

train_data = pd.read_csv(path+r'数据处理1(删除无关特征,注册日期处理)\traindata_删换版本.csv')
test_data = pd.read_csv(path+r'数据处理1(删除无关特征,注册日期处理)\test_data_删换版本.csv')

print('train data shape:',train_data.shape)
print('test data shape:',test_data.shape)

#删除多余的序号
train_data.drop(columns=['Unnamed: 0'],inplace=True)
test_data.drop(columns=['Unnamed: 0'],inplace=True)
```

# 2.构造字段‘上线时间-注册日期’  第二阶段：（汽车使用天数计算）:

```python
train_period=pd.to_datetime(train_data['汽车上线时间'].astype('str'))-pd.to_datetime(train_data['注册日期'].astype('str'))
test_period=pd.to_datetime(test_data['creatDate'].astype('str'))-pd.to_datetime(test_data['regDate'].astype('str'))
```

1. 增加'使用时间'字段：

```python
train_data['使用时间']=train_period.dt.days
col_name = train_data.columns.tolist()
print(col_name)
```

```python
train_data=train_data[['使用时间','注册日期', '车型编码', '汽车品牌', '车身类型', '燃油类型', '变速箱', '发动机功率', '已行驶公里', '有无尚未修复的损坏', '地区编码', '汽车上线时间', '交易价格', 'v_0', 'v_1', 'v_2', 'v_3', 'v_4', 'v_5', 'v_6', 'v_7', 'v_8', 'v_9', 'v_10', 'v_11', 'v_12', 'v_13', 'v_14']]
```

```python
test_data['使用时间']=test_period.dt.days
col_name = test_data.columns.tolist()
print(col_name)
```

```python
test_data=test_data[['使用时间','regDate', 'model', 'brand', 'bodyType', 'fuelType', 'gearbox', 'power', 'kilometer', 'notRepairedDamage', 'regionCode', 'creatDate', 'v_0', 'v_1', 'v_2', 'v_3', 'v_4', 'v_5', 'v_6', 'v_7', 'v_8', 'v_9', 'v_10', 'v_11', 'v_12', 'v_13', 'v_14']]
```

2. 删除‘注册日期’，‘汽车上线时间’字段：

```python
train_data=train_data.drop(['注册日期','汽车上线时间'],axis=1)
test_data=test_data.drop(['creatDate','regDate'],axis=1)
```

# 5.特征提取：

```python
train_data.corr('pearson').交易价格
```

```python
#匿名特征中相关系数小于0.2的特征进行了保留95%信息的pca降维
pca_traindata=train_data[['v_1','v_2','v_4','v_6','v_7','v_13','v_14']]
pca_testdata=test_data[['v_1','v_2','v_4','v_6','v_7','v_13','v_14']]
```

```python
from sklearn.decomposition import PCA
transfer = PCA(n_components=0.95)
traindata_new=transfer.fit_transform(pca_traindata)
testdata_new=transfer.transform(pca_testdata)
#pca降维后的数据：
pd.DataFrame(traindata_new,columns=['pca1','pca2','pca3'])
pd.DataFrame(testdata_new,columns=['pca1','pca2','pca3'])
```

```python
#合并降维后的数据和原本样本集：
train_data=pd.concat([train_data,pd.DataFrame(traindata_new,columns=['pca1','pca2','pca3'])],axis=1)
test_data=pd.concat([test_data,pd.DataFrame(testdata_new,columns=['pca1','pca2','pca3'])],axis=1)
```

```python
#删除降维的原数据：
train_data.drop(columns=['v_1','v_2','v_4','v_6','v_7','v_13','v_14'],inplace=True)
test_data.drop(columns=['v_1','v_2','v_4','v_6','v_7','v_13','v_14'],inplace=True)
```

# 算法预测：

```python
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
train_data.有无尚未修复的损坏.replace({'-':2,'0.0':0,'1.0':1},inplace=True)
test_data.notRepairedDamage.replace({'-':2,'0.0':0,'1.0':1},inplace=True)

train_data.有无尚未修复的损坏=train_data.有无尚未修复的损坏.astype('int')
test_data.notRepairedDamage=test_data.notRepairedDamage.astype('int')
```

```python
sample=train_data.drop(columns=['交易价格'])
target=train_data['交易价格']
```

```python
x_train, x_test, y_train, y_test = train_test_split(sample, target, test_size=0.1)
```

```python
estimator = XGBRegressor(n_estimators=600, learning_rate=0.1)

estimator.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
y_predict=estimator.predict(x_test)
print(mean_absolute_error(y_test,y_predict))
```

```python
#xgboost.to_graphviz(estimator, num_trees=2)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.figure(figsize=(10, 6))
plt.rcParams['figure.dpi'] = 150
xgb.plot_importance(estimator, max_num_features=7)
```

# LGB

```python
from lightgbm.sklearn import LGBMRegressor
from sklearn.metrics import mean_absolute_error
```

```python
name1={'使用时间':'usetime','车型编码':'model','汽车品牌':'brand','车身类型':'bodyType',
       '燃油类型':'fuelType','变速箱':'gearbox','发动机功率':'power','已行驶公里':'kilometer',
      '有无尚未修复的损坏':'notRepairedDamage','地区编码':'regionCode','交易价格':'prize'}
traindata = train_data.rename(columns=name1)

name2={'使用时间':'usetime'}
testdata = test_data.rename(columns=name2)
```

```python
sample=traindata.drop(columns=['prize'])
target=traindata['prize']
x_train, x_test, y_train, y_test = train_test_split(sample, target, test_size=0.2)
```

```python
estimator = LGBMRegressor(n_estimators=1200 , max_depth = 10,learning_rate=0.1)
estimator.fit(x_train,y_train)
```

```python
y_predict=estimator.predict(x_test)
print(mean_absolute_error(y_test,y_predict))
```
