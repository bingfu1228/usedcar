# 5.特征提取：(选择型)

```python
train_data.corr('pearson').交易价格
```

```python
#匿名特征中相关系数小于0.2的特征进行了保留95%信息的pca降维
pca_traindata=train_data[['v_1','v_2','v_4','v_6','v_7','v_13','v_14']]
pca_testdata=test_data[['v_1','v_2','v_4','v_6','v_7','v_13','v_14']]
```

```python
from sklearn.decomposition import PCA
transfer = PCA(n_components=0.95)
traindata_new=transfer.fit_transform(pca_traindata)
testdata_new=transfer.transform(pca_testdata)
#pca降维后的数据：
pd.DataFrame(traindata_new,columns=['pca1','pca2','pca3'])
pd.DataFrame(testdata_new,columns=['pca1','pca2','pca3'])
```

```python
#合并降维后的数据和原本样本集：
train_data=pd.concat([train_data,pd.DataFrame(traindata_new,columns=['pca1','pca2','pca3'])],axis=1)
test_data=pd.concat([test_data,pd.DataFrame(testdata_new,columns=['pca1','pca2','pca3'])],axis=1)
```

```python
#删除降维的原数据：
train_data.drop(columns=['v_1','v_2','v_4','v_6','v_7','v_13','v_14'],inplace=True)
test_data.drop(columns=['v_1','v_2','v_4','v_6','v_7','v_13','v_14'],inplace=True)
```
